1. Escriba un código que sume los números múltiplos de 5 y 17 que hay entre 0 y 100,000, es decir, que sume 0 + 5 + 10 + 17 + 20 + 25 + 30 + 34 + 35 + 51 + ...

````python
suma = 0
for n in range(0,100001):
    if n % 5 == 0 or n % 17 == 0:
        suma += n
print("La suma es:", suma)
````

2. Haga la gráfica de la función $f(x)=x - \frac{x^2}{2} + \frac{x^3}{3}$ y de la función $\ln (1+x)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo $|x|<1$. Explique las similitudes y diferencias.

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x - x**2/2 + x**3/3

def g(x):
    return np.log(1+x)
````

````python
x = np.linspace(-0.9,0.9,1000)
````

````python
plt.plot(x,f(x),label="f(x) = x - x^2/2 + x^3/3")
plt.plot(x,g(x),label="ln(1+x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
````

. Realice una función que utilice la función rand() para simular un dado de 7 caras. Es decir, haga una función que proporcione un número entre 1 y 8. Por ejemplo:
```python
dado()
```
y a la salida, podría proporcionar el número: ```5```.

````python
import numpy as np

def dado():
    r = np.random.rand()
    resultado = int(7*r) + 1
    return resultado

print(dado())
````

4. Realice una función que reciba un número de cuenta de exlusivamente 9 dígitos, y que a la salida proporcione la suma y el producto de todos de todos sus dígitos distintos de cero. Compruebe el funcionamiento de su función con algún número de cuenta. Ejemplos:
```
In: sumar_nc(456982671)
Out: Suma = 48 ; Producto = 725,760
```

```
In: sumar_nc(623548)
Out: El número no tiene 9 dígitos
```

````python
def sumar_nc(numero):
    if len(str(numero)) != 9:
        print("El numero no tiene 9 dígitos")
        return
    suma = 0
    producto = 1
    for dig in str(numero):
        if dig != '0':
            d = int(dig)
            suma += d
            producto *= d
    print("Suma =", suma, "; Producto =", producto)
````

````python
sumar_nc(456982271)
````
