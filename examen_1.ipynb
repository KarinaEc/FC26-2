2. Haga la gráfica de la función $f(x)=x^2 - \frac{x^4}{2} + \frac{x^6}{3} - \frac{x^8}{4} + \frac{x^{10}}{5} $ y de la función $\ln (1+x^2)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo $|x|<1$. Explique las similitudes y diferencias.

*(2 puntos)*

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - x**4/2 + x**6/3 - x**8/4 + x**10/5 

def g(x):
    return np.log(1+x**2)

x = np.linspace(-0.9,0.9,1000)
````

````python
plt.plot(x,f(x),label="f(x) = x**2 - x**4/2 + x**6/3 - x**8/4 + x**10/5")
plt.plot(x,g(x),label="ln(1+x**2)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
````

Las 2 funciones son casi identicas sólo cambia que log(1+x^2) es la función exacta mientras que f(x) es una aproximación.

4. Realice una función que reciba una fecha en formato DDMMAAAA (8 dígitos), y que a la salida proporcione la suma y el producto de todos de todos sus dígitos distintos de cero. Compruebe el funcionamiento de su función con alguna fecha. Ejemplos:
```
In: sumar_numero(12032026)
Out: Suma = 16 ; Producto = 144
```

```
In: sumar_nc(311986)
Out: El número no tiene 9 dígitos
```

*(4 puntos)*

````python
def sumar_nc(numero):
    if len(str(numero)) != 8:
        print("El numero no tiene 8 dígitos")
        return
    suma = 0
    producto = 1
    for dig in str(numero):
        if dig != '0':
            d = int(dig)
            suma += d
            producto *= d
    print("Suma =", suma, "; Producto =", producto)
````

````python
sumar_nc(12345678)
````

````python
sumar_nc(1234567)
````
