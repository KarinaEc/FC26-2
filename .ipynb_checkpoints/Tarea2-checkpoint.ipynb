## Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt

def sistema_ejercicio1(t, y):
    y1, y2 = y
    dy1 = y2
    dy2 = np.exp(-t)*(np.sin(3*t) + np.cos(2*t)) - 0.5*y2 - np.sin(y1)
    return np.array([dy1, dy2])

# a) Método de Euler
def euler_step(f, t, y, h):
    return y + h * f(t, y)

# b) Método de Runge-Kutta de 2do orden
def rk2_step(f, t, y, h):
    k1 = f(t, y)
    k2 = f(t + 0.5*h, y + 0.5*h*k1)
    return y + h * k2

# c) Método de Runge-Kutta de 4to orden
def rk4_step(f, t, y, h):
    k1 = f(t, y)
    k2 = f(t + 0.5*h, y + 0.5*h*k1)
    k3 = f(t + 0.5*h, y + 0.5*h*k2)
    k4 = f(t + h, y + h*k3)
    return y + (h/6.0) * (k1 + 2*k2 + 2*k3 + k4)
```

```python
def integrar(metodo, f, t_span, y0, h):
    t_values = np.arange(t_span[0], t_span[1], h)
    y_values = np.zeros((len(t_values), len(y0)))
    y_values[0] = y0
    
    for i in range(len(t_values) - 1):
        y_values[i+1] = metodo(f, t_values[i], y_values[i], h)
        
    return t_values, y_values
```

```python
t_span = (0, 20)  
h = 0.05          
y0 = [0, 1]       

t_eu, y_eu = integrar(euler_step, sistema_ejercicio1, t_span, y0, h)
t_rk2, y_rk2 = integrar(rk2_step, sistema_ejercicio1, t_span, y0, h)
t_rk4, y_rk4 = integrar(rk4_step, sistema_ejercicio1, t_span, y0, h)

plt.figure(figsize=(10, 6))
plt.plot(t_eu, y_eu[:, 0], label='Euler', linestyle='--')
plt.plot(t_rk2, y_rk2[:, 0], label='RK2', linestyle='-.')
plt.plot(t_rk4, y_rk4[:, 0], label='RK4', linewidth=2)
plt.title('Solución $x(t)$ para la Ecuación Diferencial')
plt.xlabel('Tiempo (t)')
plt.ylabel('Posición (x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Ejercicio 2

```python
def sistema_ej2(t, u):
    
    y, dy, d2y = u
    
    du0 = dy
    du1 = d2y
    du2 = np.exp(-t) * np.sin(3*t) - d2y**2 + 3*dy - np.cos(y)**2
    
    return np.array([du0, du1, du2])
```

```python
def rk4_integrador(f, t_span, u0, h):
    t = np.arange(t_span[0], t_span[1] + h, h)
    u = np.zeros((len(t), len(u0)))
    u[0] = u0
    for i in range(len(t) - 1):
        k1 = f(t[i], u[i])
        k2 = f(t[i] + h/2, u[i] + h/2 * k1)
        k3 = f(t[i] + h/2, u[i] + h/2 * k2)
        k4 = f(t[i] + h, u[i] + h * k3)
        u[i+1] = u[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    return t, u

u0 = [1.0, 2.0, 1.0] 
t_intervalo = (1.0, 1.65)

h_grande = 0.001
h_chico = 0.0001

t1, sol1 = rk4_integrador(sistema_ej2, t_intervalo, u0, h_grande)
t2, sol2 = rk4_integrador(sistema_ej2, t_intervalo, u0, h_chico)

print(f"Resultado en t=1.65 con h={h_grande}: {sol1[-1, 0]:.6f}")
print(f"Resultado en t=1.65 con h={h_chico}: {sol2[-1, 0]:.6f}")
```

```python
t_ext, sol_ext = rk4_integrador(sistema_ej2, (1.0, 1.8), u0, 0.0001)

plt.figure(figsize=(10, 5))
plt.plot(t_ext, sol_ext[:, 0], label='Posición $y(t)$')
plt.axvline(x=1.65, color='r', linestyle='--', label='Límite solicitado')
plt.xlabel('Tiempo $t$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.show()
```

Al intentar integrar hacia $t=2$, el sistema experimenta un crecimiento explosivo. Esto se debe a la naturaleza no lineal de la ecuación, específicamente al término $\ddot{y}^2$.  
A medida que la aceleración aumenta, la tercera derivada se ve fuertemente afectada, lo que impulsa la velocidad y la posición hacia el infinito en un tiempo finito    

## Ejercicio 3  

```python
def rk4_step(f, t, u, h, *args):
    k1 = f(t, u, *args)
    k2 = f(t + 0.5*h, u + 0.5*h*k1, *args)
    k3 = f(t + 0.5*h, u + 0.5*h*k2, *args)
    k4 = f(t + h, u + h*k3, *args)
    return u + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)

def rk4_integrador(f, t_span, u0, h, *args):
    t = np.arange(t_span[0], t_span[1] + h, h)
    u = np.zeros((len(t), len(u0)))
    u[0] = u0
    for i in range(len(t) - 1):
        u[i+1] = rk4_step(f, t[i], u[i], h, *args)
    return t, u
```

Definimos la ecuación: $\ddot{x} + 0.1\dot{x} + 4\sin x = 0.5\sin(\omega t)$.

```python
def sistema_ej3(t, u, w):
    x, vx = u
    dx = vx
    dvx = 0.5 * np.sin(w * t) - 0.1 * vx - 4 * np.sin(x)
    return np.array([dx, dvx])
```

```python
t_span = (0, 40) 
u0 = [0.0, 1.0]  # x(0)=0, v(0)=1
h = 0.01
frecuencias = [1.0, 1.5, 2.0, 2.5, 3.0] 

plt.figure(figsize=(12, 6))

for w in frecuencias:
    t, sol = rk4_integrador(sistema_ej3, t_span, u0, h, w)
    plt.plot(t, sol[:, 0], label=f'$\omega = {w}$')

plt.xlabel('Tiempo $t$')
plt.ylabel('Posición $x(t)$')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

El sistema tiene una frecuencia natural aproximada de $\omega_0 = \sqrt{g/L} = \sqrt{4} = 2$. Cuando $\omega$ se acerca a 2, la amplitud de las oscilaciones aumenta significativamente.  
Para frecuencias muy bajas ($\omega < 2$), el sistema suele seguir la fuerza externa. Para frecuencias muy altas ($\omega > 2$), la respuesta del sistema es más pequeña debido a la inercia y el amortiguamiento.  

## Ejercicio 4

```python
def lorenz_system(t, state):
    x, y, z = state
    dx = 10 * (y - x)
    dy = x * (28 - z) - y
    dz = x * y - (8/3) * z
    return np.array([dx, dy, dz])
```

## a)

```python
u0_rand = np.random.uniform(-5, 5, 3)
t_span = (0, 50) 
h = 0.01

t, sol = rk4_integrador(lorenz_system, t_span, u0_rand, h)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plano xy
axs[0].plot(sol[:, 0], sol[:, 1], color='blue', lw=0.5)
axs[0].set_title('Plano XY')
axs[0].set_xlabel('x'); axs[0].set_ylabel('y')

# Plano xz
axs[1].plot(sol[:, 0], sol[:, 2], color='red', lw=0.5)
axs[1].set_title('Plano XZ')
axs[1].set_xlabel('x'); axs[1].set_ylabel('z')

# Plano yz
axs[2].plot(sol[:, 1], sol[:, 2], color='green', lw=0.5)
axs[2].set_title('Plano YZ')
axs[2].set_xlabel('y'); axs[2].set_ylabel('z')

plt.tight_layout()
plt.show()
```

En las gráficas de los planos (especialmente el XZ), se observan dos "alas".  
El sistema orbita alrededor de un centro y luego salta impredeciblemente al otro. Aunque el movimiento parece azaroso, está confinado a esta estructura geométrica específica llamada atractor extraño.

## b)

```python
u0_1 = np.array([1.0, 1.0, 1.0])
u0_2 = np.array([1.0, 1.01, 1.0])  # Difieren en 10^-2 en 'y'
u0_3 = np.array([1.0, 1.02, 1.0])

t_span_caos = (0, 40)

t1, sol1 = rk4_integrador(lorenz_system, t_span_caos, u0_1, h)
t2, sol2 = rk4_integrador(lorenz_system, t_span_caos, u0_2, h)
t3, sol3 = rk4_integrador(lorenz_system, t_span_caos, u0_3, h)

plt.figure(figsize=(12, 6))
plt.plot(t1, sol1[:, 0], label='Trayectoria 1')
plt.plot(t2, sol2[:, 0], label='Trayectoria 2', linestyle='--')
plt.plot(t3, sol3[:, 0], label='Trayectoria 3', linestyle=':')
plt.title('Sensibilidad a las condiciones iniciales (x)')
plt.xlabel('Tiempo')
plt.ylabel('x')
plt.legend()
plt.grid(True)
plt.show()
```

Durante los primeros segundos ($t < 15$ aprox.), las tres líneas están encimadas. Pero despues de cierto tiempo, las trayectorias se separan por completo.

Este comportamiento demuestra que en sistemas caóticos, la predicción a largo plazo es imposible, ya que cualquier error infinitesimal en la medición inicial (o en el redondeo numérico) se amplifica exponencialmente con el tiempo.
