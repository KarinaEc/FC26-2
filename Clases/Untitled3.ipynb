```python
import folium
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,IFrame
```

```python
mis_coordenadas = [19.323644794070223, -99.17898375489459]
mapa = folium.Map(location=mis_coordenadas,zoom_start=15)
folium.Marker(mis_coordenadas, popup='Aqui toy').add_to(mapa)
mapa.save('./DatosCSV/Mapita.html')
```

```python
mis_datos = np.genfromtxt('./DatosCSV/UbiCasa.csv',delimiter=',',dtype=float)
t = mis_datos[1:,0]
latitud = mis_datos[1:,1]
lat_long = mis_datos[1:,1:3]
mi_mapita = folium.Map(location=lat_long[0], zoom_start=15) #solo acpeta un punto, si se pone mas puntos los pone como pines
trayectoria = folium.PolyLine(locations=lat_long,color="#C8098B",weight=6).add_to(mi_mapita)
folium.Marker(location=lat_long[0],popup='Start',icon=folium.Icon(color="gray")).add_to(mi_mapita)
folium.Marker(location=lat_long[-1],popup='End',icon=folium.Icon(color="lightblue")).add_to(mi_mapita)
mi_mapita.save('MapitaKari.html')
IFrame(src='MapitaKari.html',width=350,height=350)
```
